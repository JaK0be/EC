### Parâmetros

```python
# Tabela de valores (página 29) para p-I
"""
q - Modulus
Le - Bound constant for error polynomials
Ls - Bound constant for the secret polynomial
E - Rejection bound used (for signing and verification) related to Le
S - Rejection bound used (for signing and verification) related to Ls
B - Bound for the random polynomial at signing
d - Rounding value
"""
k = 256
kk = 4
bgen_a = 108
rate_xof = 168
q = 343576577
n = 1024
beta = 64 # Targeted precision: Tabela 1, página 20
chunk_size = 512 # c -> Implementation details, página 20
t = 78 # number of rows of cdt table
h = 25
Ls = 554
Le = 554
B = 2**19 - 1
d = 22
```

### Funções auxiliares

```python
import hashlib
from CompactFIPS202 import SHAKE128, Keccak
from cshake import cSHAKE128

# "PRF1: the XOF is instantiated with SHAKE128 (resp. SHAKE256) for the Level-I parameter set (resp. for the Level-III parameter set); see Algorithm 9"
# seeds: s, e1, e2, e3, e4, a, y
def prf1(pre_seed):
    # k*(kk + 3)/8) = 256
    seed = hashlib.shake_128(pre_seed).digest(int(k*(kk + 3)/8))
    
    seed_s = seed[0:32]
    seed_e1 = seed[32:64]
    seed_e2 = seed[64:96]
    seed_e3 = seed[96:128]
    seed_e4 = seed[128:160]
    seed_a = seed[160:192]
    seed_y = seed[192:224]
    
    return seed_s, seed_e1, seed_e2, seed_e3, seed_e4, seed_a, seed_y

def prf2(seed_y, r, digest):
    inp = seed_y + r + digest
    res = hashlib.shake_128(inp).digest(int(k/8))
    return res

def genA(seed_a):
    b = int(round(math.log(q,2)/8))
    D = 0
    b1 = bgen_a
    c_temp = cSHAKE128(seed_a, rate_xof*b1, b'0')
    tam = int((rate_xof*b1) / (b*8))
    c = [0] * tam
    a = [[0 for i in range(n)] for j in range(kk+1)]
    for i in range(tam):
        c[i] = c_temp[i*b:(i+1)*b]
    #print(c)
    i = 0
    pos = 0
                
    while i<(kk*n):
        if pos > int((math.floor((rate_xof*b1)/(b*8))) - 1):
            D = D+1
            pos = 0
            b1 = 1
            h = cSHAKE128(seed_a, rate_xof*b1, int(D).to_bytes(2, byteorder="big"))
            for j in range(tam):
                c[j] = h[j*b:(j+1)*b]
        ph = int.from_bytes(c[pos], "big") % 2**(int(round(math.log(q,2))))
        if ph < q:
            hor = int(math.floor(i/n)) + 1
            ver = i - n*int(math.floor(i/n))
            a[hor][ver] = ph
            i = i + 1
        pos = pos + 1    
    
    a.pop(0)
    return a

def gaussSampler(seed, D):
    D1 = D * 2**8
    z = [0] * n
    r = [0] * int((chunk_size*beta/8))
    i = 0
    for i in range(n):
        r_temp = cSHAKE128(seed, chunk_size*beta/8, int(D1).to_bytes(3, byteorder="big"))
        for v in range(chunk_size*beta/8):
            r[v] = r_temp[v*(beta/8):(1+v)*(beta/8)]
        D1 = D1 + 1
        for j in range(chunk_size):
            z[i+j] = 0
            rj = int.from_bytes(r[j], "big")
            sign = rj/2**(beta-1)
            val = rj
            for kp in range(t):
                if val >= cdt_v[kp]:
                    z[i+j] = z[i+j] + 1
            if sign == 1:
                z[i+j] = -z[i+j]
        i = i + chunk_size
    return z

# "XOF is given by XOF(X,L,D), where X is the input string, L specifies the output length in bytes, and D specifies an optional domain separator"
def ySampler(rand, D):
    b = int(round((math.log(B,2)+1)/8))
    pos = 0
    nl = n
    Dl = D * (2**8)
    c = [0] * nl
    y = [0] * n
    c_temp = cSHAKE128(rand, b*nl, int(Dl).to_bytes(2, byteorder="big"))
    for v in range(nl):
        c[v] = c_temp[v*b:(v+1)*b]
    while i < n:
        if pos >= nl:
            Dl = Dl + 1
            pos = 0
            nl = int(math.floor(rate_xof/b))
            c_t = cSHAKE128(rand, rate_xof, int(Dl).to_bytes(2, byteorder="big"))
            for v in range(nl):
                c[v] = c_t[v*b:(v+1)*b]
        y[i] = int.from_bytes(c[pos],"big") % 2**(int(round(math.log(B,2)))+1) - B
        if y[i] != (B + 1):
            i = i + 1
        pos = pos + 1
    return y

def checkS(s):
    ls = list(s)
    ls.sort(reverse=True)
    sum = 0
    for i in range(h):
        sum += ls[i]
    if sum > Ls:
        return 1
    return 0

def checkE(e):
    ls = list(e)
    ls.sort(reverse=True)
    sum = 0
    for i in range(h):
        sum += ls[i]
    if sum > Le:
        return 1
    return 0


def mulPoly(e, f, mod):
        h = [0]*kk
        for i in range(0, kk):
            h[i] = (e[i]*f[i]) % mod
        return h
    
def H(v, gm, g):
    for i in range(1,(k+1)):
        for j in range(n):
            val = v[i][j] % 2**d
            if val > 2**(d-1):
                val = val - 2**d
            w[(i-1)*n+j] = (v[i][j] - val)/2**d
    w = [0] * (k*n+80)
    for ca in range(40):
        w[kk*n+ca] = gm[i]
        w[kk*n+40+ca] = g[i]
    cl = hashlib.shake_128(w).digest(int(k/8))
    return cl

def enc(cl):
    D = 0
    cnt = 0
    r = cSHAKE128(cl, rate_xof, int(D).to_bytes(2, byteorder="big"))
    i = 0
    c = [0] * n
    pos_list = [0] * h
    sign_list = [0] * h
    while i < h:
        if cnt > (rate_xof - 3):
            D = D + 1
            cnt = 0
            r = cSHAKE128(cl, rate_xof, int(D).to_bytes(2, byteorder="big"))
        pos = (r[cnt] * 2**8 + r[cnt+1]) % n
        if c[pos] == 0:
            if (r[cnt+2] % 2) == 1:
                c[pos] = -1
            else:
                c[pos] = 1
            pos_list[i] = pos
            sign_list[i] = c[pos]
            i = i + 1
        cnt = cnt + 3
    return (pos_list, sign_list)

def mod(val, q):
    cl = val % q
    if cl > int(math.floor(q/2)):
        cl = cl - q
    return cl
        
# we define the infinity norm for a polynomial ||f||oo = max||f[i]||oo
# the function max(f) returns the k-th largest absolute coeficient of f
def infinity_norm(val):
    tam = len(val)
    temp = 0
    for i in range(tam):
        if abs(f[i]) > temp:
            temp = abs(f[i])
    return temp    
```

### Geração de chaves, assinatura e verificação
#### Geração de chaves
- Primeiro são gerados os polinómios *a1...ak*, de forma uniforme e aleatória usando.
    - Para isso usa-se `prf1` para expandir uma `pre_seed`. Um dos resultados dessa expansão será `seed_a` que será usado para como input de `genA` que gera os polinómios *a1...ak* e guarda-os em `a`.
- É escolhido polinómio `s` com uma distribuição Gaussiana.
    - O polinómio deve passar o teste `checkS`.
- O passo anterior é repetido para escolher os polinómios secretos de erro e1...ek, mas neste caso devem passar o teste `checkE`.
- Calcular o polinómio *t = a * s + e mod q*, que faz parte da chave pública e a sua hash: *g = G(t)*.
- Retornar *sk = (`s`, `e`, `seed_a`, `seed_y`, `g`)* e *pk = (`t`, `seed_a`)*.

#### Assinatura
- Gerar um `y` uniformemente de forma aleatória, usando `ySampler`.
    - Para fornecer aleatoriedade é utilizada uma string aleatória `rand`, que é calculada usando `prf2(seed_y, r, G(m))`, sendo que `r` = é uma string aleatória e `G(m)` é o resultado de fazer hashing da mensagem.
- A `seed_a` é expandida para gerar os polinómios *a1...ak*, que são utilizados para calcular os polinómios *v1...vk*, sendo que $v[i] = a[i] * y$ $mod±q$.
- Calcular *cl = H(v, G(m), g)*.
- Gerar o polinómio pseudo-aleatório c que é dividido em 2 arrays, `pos_list` e `sign_list`que representam as posições e sinais dos coeficientes diferentes de zero de *c*, respetivamente.
- Calcular *z = y + s * c*, que faz parte da assinatura.
    - *z* deve passar passar os testes de segurança e correção.
    - Caso falhe algum dos testes *z* e *cl* são descartados e o algoritmo é reiniciado a partir da geração de `y`.
- Retornar assinatura *(z,cl)*.

#### Verificação
- Calcular c = (`pos_list`, `sign_list`) a partir de *cl* presente na assinatura.
- Expandir a `seed_a` (presente na pk) para gerar os polinómios *a1...ak* usando `genA`.
- Calcular *w1...wk*.
    - $w[i] = a[i] * z - t[i] * c$ $mod±q$
- Calcular *H(w, G(m), G(t))*.
    - Se este resultado for igual a *cl* e *z* passar o teste de segurança então a assinatura é válida.

```python
import os, random

# "XOF is given by XOF(X,L,D), where X is the input string, L specifies the output length in bytes, and D specifies an optional domain separator"
def keyGen():
    counter = 1
    e = [0] * kk
    t = [0] * kk
    pre_seed = os.urandom(32)
    seed_e = [0] * kk
    seed_s, seed_e[0], seed_e[1], seed_e[2], seed_e[3], seed_a, seed_y = prf1(pre_seed)
    a = genA(seed_a)
    s = gaussSampler(seed_s, counter)
    counter = counter + 1
    bv = checkS(s)
    while bv != 0:
        s = gaussSampler(seed_s, counter)
        counter = counter + 1
        bv = checkS(s)
    for i in range(kk):
        while True:
            e[i] = gaussSampler(seed_e[i], counter)
            counter = counter + 1
            if checkE(e[i]) == 0:
                break
        #----------------------------
        t[i] = (a[i]*s + e[i]) % q
        #----------------------------

    inp = b''.join(t)
    g = hashlib.shake_256(inp).digest(int(40))
    
    sk = (s, e[0], e[1], e[2], e[3], seed_a, seed_y, g)
    pk = (t[0], t[1], t[2], t[3], seed_a)
    
    return sk, pk

def sign(m, sk):
    E = Le
    counter = 1
    r = os.urandom(32)
    seed_y = sk[6]
    seed_a = sk[5]
    g = sk[7]
    s = sk[0]
    rand = prf2(seed_y, r, hashlib.shake_256(m).digest(int(40)))
    a = genA(seed_a)
    while True:
        y = ySampler(rand, counter)
        v = [[0 for i in range(n)] for i in range(k)]
        z = [0] * n
        w = [[0 for i in range(n)] for i in range(k)]
        for i in range(kk):
            v[i] = [0 for j in range(n)]
            for j in range(n):
                v[i][j] = mod((a[i][j] * y[j]), q)
        cl = H(v, hashlib.shake_256(m).digest(int(40)), g)
        (pos_list, sign_list) = enc(cl) # <---- c
        #---------------------------------
        z = y + s*c # sendo que c é o resultado de enc(cl)
        #---------------------------------
        flag = True
        for i in range(n):
            if z[i] < -B or z[i] > B:
                counter = counter + 1
                flag = False
                break
                
        if flag == False:
            continue
            
        flagl = True
        for i in range(1, kk+1):
            #----------------------------------
            w[i] = v[i] - e[i]*c % q # mod e poly
            #----------------------------------
            # Os valores de w já tem o mod+-q feito acima
            val = [mod(w[i][pl], 2**d) for pl in range(n)]
            if (infinity_norm(val) >= 2**(d-1)-E) or (infinity_norm(w[i]) >= (int(math.floor(q/2)) - E)):
                counter = counter + 1
                flag = False
                break
        if flagl == True:
            return (z, cl)

def verify(m, signature, pk):
    (z, cl) = signature
    seed_a = pk[4]
    
    (pos_list, sign_list) = enc(cl)
    a = genA(seed_a)
    for i in range(kk):
        #----------------------------
        w[i] = a[i]*z - t[i]*c % q # <------ mod, poly
        #----------------------------
    
    inp = b''.join(t)
    gm = hashlib.shake_256(m).digest(int(40))
    gt = hashlib.shake_256(inp).digest(int(40))
    h_res = H(w, gm, gt)
    
    if cl != h_res:
        return -1
    
    for i in range(n):
        if z[i] < -B or z[i] > B:
            return -1
    return 0
    
#sk, pk = keyGen()
```

### Script de geração da tabela *cdt_v*
O script presente na secção abaixo é fornecido no pacote de submissão do qTesla e é utilizado para gerar a tabela *cdt_v* utilizada em `gaussSampler`.

```python
# CDT table generation for a given sigma and targeted precision beta
# Tables are generated assuming a wordsize of 32 bits, i.e., prec=32

import time


def rho(k, sigma):
    return exp(-(k/sigma)^2/2)


hexdigits = "0123456789ABCDEF"
def hexstring(w, n):
    global hexdigits
    s = ""
    while w > 0:
        s = hexdigits[w & 15] + s
        w >>= 4
    s = ("0"*n)[0:n-len(s)] + s
    return s


# half-table with ((prec-1)*t + 1)-bit sorting keys:
def makeCDT(sigma, border, t, prec):
    print("/*************************************************************************************")
    print("* qTESLA: an efficient post-quantum signature scheme based on the R-LWE problem")
    print("*")
    print("* Abstract: CDT constants for the Gaussian sampler")
    print("**************************************************************************************/")
    print("")
    print("#ifndef CDTSAMP" + str(prec))
    print("#define CDTSAMP" + str(prec))
    print("")
    print("#include <stdint.h>")
    print("#include \"params.h\"")
    print("")
    print("// sigma =", float(sigma), ",", prec*t, "-bit precision")
    # compute normalization factor:
    S = rho(0, sigma)
    for k in range(1, border + 1):
        S += 2*rho(k, sigma)
    #print "// S =", float(S)
    print("")
    print("#define CDT_ROWS " + str(border + 1))
    print("#define CDT_COLS " + str(t))
    print("")
    print("static const int" + str(prec) + "_t cdt_v[CDT_ROWS*CDT_COLS] = {")

    suffix = "LL" if prec == 64 else "L" if prec == 32 else ""
    line = "    "
    for c in range(t - 1, -1, -1):
        line += "0x" + hexstring(0, prec >> 2) + suffix + ", "
    line += "// 0"
    print(line)

    shift = 1 << ((prec - 1)*t)
    v = rho(0, sigma)/S

    w = floor(v*shift)
    line = "    "
    for c in range(t - 1, -1, -1):
        d = w >> ((prec - 1)*c)
        w -= d << ((prec - 1)*c)
        line += "0x" + hexstring(d, prec >> 2) + suffix + ", "
    line += "// 1"
    print(line)

    ok = True
    for k in range(1, border):
        v += 2*rho(k, sigma)/S

        w = floor(v*shift)
        line = "    "
        for c in range(t - 1, -1, -1):
            d = w >> ((prec - 1)*c)
            w -= d << ((prec - 1)*c)
            line += "0x" + hexstring(d, prec >> 2) + suffix + ", "
        line += "// " + str(k + 1)
        print(line)

        if d == ((1 << (prec - 1)) - 1):
            if k + 1 != border:
                print("// ----------------------------------------------------")
                print("// >>> PLEASE RUN THE SCRIPT AGAIN WITH border =", k + 1, "<<<")
                ok = False
                break
        else:
            if k + 1 == border:
                print("// ----------------------------------------------------")
                print("// >>> PLEASE RUN THE SCRIPT AGAIN WITH A LARGER border <<<")
                ok = False
                break
    print("}; // cdt_v")
    print("")
    if ok:
        #print "// actual precision:", float(-log(abs(S - v), 2)), "bits"
        print("// memory requirements:")
        for k in range(11, 4, -1):
            s = str(1 << k)
            s = (" "*4)[0:4-len(s)] + s
            u = str(((1 << k) + border + 1)*(t + 1)*(prec >> 3))
            u = (" "*5)[0:5-len(u)] + u
            print("//    ", s, "samples:", u, "bytes")
        print("// table alone:", (prec >> 3)*(border + 1)*t, "bytes")
        print("")
    print("#endif // CDTSAMP" + str(prec) + "\n")


flavor = "qTESLA-p-I"
#flavor = "qTESLA-p-III"

prec = 32

elapsed = -time.clock()
if flavor == "qTESLA-p-I":
    sigma = ceil(100* 8.50)/100
    if prec == 32:
        makeCDT(sigma,  77, 2, 32)   #  beta = 64-bit samples
    else:
        makeCDT(sigma,  78, 1, 64)   #  beta = 64-bit samples
elif flavor == "qTESLA-p-III":
    sigma = ceil(100* 8.50)/100
    if prec == 32:
        makeCDT(sigma, 110, 4, 32)   # beta = 128-bit samples
    else:
        makeCDT(sigma, 111, 2, 64)   # beta = 128-bit samples
else:
    print("NOT IMPLEMENTED")
elapsed += time.clock()

#print ""
#print ">>> all done:", elapsed, "s."
```
